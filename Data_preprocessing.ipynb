 # Random sample imputation

###
* Sample function to randomly select 50 values from the column Age
* after removinf null values
* random_state: reproducing the same result u can assign any integer to it.
* When we do not mention random_state, the sample function will select any
  3 values randomly from the column
* Everytime you run the program the values choosen by sample function 
  will change
* This will generate different result everytime
* To avoid this we mentioned the random_state as a fixed number, so that
  everytime when the program is executed we get the same result. This is 
  called reproducability of the result.
* In the below code , null values are dropped, from the rest of the total 
 values present in that column will be filled.
* we can any integer value to fix the random_state for ex: random_state=5 or 20
    any integer.

```python
b=df['Age'].isnull().sum()
b
```

```python
random_sample=df['Age'].dropna().sample(b,random_state=0)
random_sample
```

```python
df['Age'].nunique()
```

```python
# The index of these random values should matched with the null values of the column
random_sample.index=df[df['Age'].isnull()].index
random_sample
```

```python
# Now finally in those index values, the samples values so chosen are filled 
df['Age']=df['Age'].fillna(random_sample)
```

# Detecting Outliers

```python
tips_df=sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
def iqr_func(data):
    q1=np.percentile(data,25,interpolation='midpoint')
    q2=np.percentile(data,50,interpolation='midpoint')
    
    q3=np.percentile(data,75,interpolation='midpoint')
    iqr=q3-q1
    lo=q1-(1.5*iqr)
    up=q3+(1.5*iqr)
    outlier=[]
    for x in data:
        if((x>up) or (x<lo)):
            outlier.append(x)
    print('outlier in the dataset are:',outlier)
iqr_func(tips_df['total_bill'])
            
    
    
```

```python
# Z score method for detecting outlier
def norm_fun(data):
    outlier=[]
    mean=np.mean(data)
    std=np.std(data)
    for x in data:
        z=(x-mean)/std
        if z>3 or z<-3:
            outlier.append(x)
            
    print('outlier of the dataset are:',outlier)
norm_fun(tips_df['total_bill'])
```
